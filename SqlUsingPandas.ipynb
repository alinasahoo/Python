# **GROUP BY / COUNT / ORDER BY**

```python
# select type, count(id) from airports group by type

airports.groupby('type').agg(count_column=('id', 'count')) # .agg(name_of_col=('on_which_col_is_agg_applied', 'name_of_agg'))

# select type, count(id) from airports group by type having count(*) > 100

x = airports.groupby('type').agg(count_column=('id', 'count'))
# GROUP BY, HAVING - in a diff step

y = x[x.count_column > 100]

# select type, count(id) from airports group by type having count(*) > 100 ORDER BY count_column DESC, type ASC

z = y.sort_values(['count_column', 'type'], ascending=[False, True])

z

# GROUP BY, HAVING, ORDER

# airports.set_index('id')

# count  = total no of non null values
# size = total no of values including missing or null values,
# min
# max
# sum

# Use size when you need to know the total number of elements, including missing or null values,
# and use count when you need to know the number of non-null values in your data
```

```python
# select iso_country, type, count(*) from airports group by iso_country, type order by iso_country, type

# airports.groupby(['iso_country', 'type']).agg(count_column=('id', 'count'))

airports.groupby(['iso_country', 'type']).size() # This returns a series # DON"T READ THIS
```

```python
# select iso_country, type, count(*) from airports group by iso_country, type order by iso_country, count(*) desc

airports.groupby(['iso_country', 'type']).agg(count=('iso_country', 'count')).sort_values(['iso_country', 'count'], ascending=[True, False]).reset_index()
```

# **HAVING**

```python
tab = {'name': ["A", "B", "C", "D"],
       'age': [10, 20, 30, 16],
       'salary': [1000, 20000, 9500, 7000],
       'dept': ['IT', 'IT', 'Sales', 'Marketing']}

df = pd.DataFrame(tab)

# select min(age), max(salary), dept FROM Emp GROUP by dept
df.groupby('dept').agg(min_age=('age', 'min'), max_salary=('salary', 'max'))
# We can write having in the next step
```

```python
# SELECT dept, min_age, max_salary from Employee GROUP BY dept HAVING MIN(age) < 20 AND max_salary > 6000

# GROUP BY ke baad sab ek alag table mai likhlo fir .reset_index() karo fir apply filtering, sorting and limit.

df.reset_index() # Brings everything to same level

a = df.groupby('dept').agg(min_age=('age','min'), max_salary=('salary', 'max'))

a[(a.min_age < 20) & (a.max_salary > 6000)].reset_index()
```

# **Aggregate functions (MIN, MAX, MEAN)**

# **JOIN**

```python
emp = {'id': [10, 20, 30, 40, 50, 70],
       'name': ["A", "B", "C", "D", "E", 'G'],
       'age': [10, 20, 10, 16, 70, 102],
       'dept': ['IT', 'IT', 'Sales', 'Marketing', 'Marketing', 'Sales']}
emp_df = pd.DataFrame(emp)

sal = {'id': [10, 20, 30, 40, 50, 60],
       'name': ["A", "B", "C", "D", "E", "F"],
       'salary': [1000, 2000, 3000, 10000, 100, 1500]}
sal_df = pd.DataFrame(sal)

# df1.merge(df2)

# select id, name, dept, salary FROM emp JOIN sal ON emp.id = sal.id WHERE age > 10 AND sal > 100

# A JOIN B ON a.col1 = B.col2 left_on = 'col1', right_on='col2'
# A JOIN B ON a.col = B.col left_on = 'col', right_on='col' OR on='col'
# how = inner, outer, left, right
emp_df[emp_df.age > 10].merge(sal_df[sal_df.salary > 100], left_on='id', right_on='id', how='outer')[['id', 'name_x', 'dept', 'salary']]
```

# **UNION / UNION ALL**

```python
emp_1 = {'id': [10, 20, 30, 40, 50, 70],
       'name': ["A", "B", "C", "D", "E", 'G']}
emp_1_df = pd.DataFrame(emp_1)

emp_2 = {'id': [10, 20, 33, 43, 53, 73],
       'name': ["A", "B", "H", "I", "J", 'L']}
emp_2_df = pd.DataFrame(emp_2)

pd.concat([emp_1_df, emp_2_df]) # UNION
```

```python
emp_1 = {'id': [10, 20, 30, 40, 50, 70],
       'name': ["A", "B", "C", "D", "E", 'G']}
emp_1_df = pd.DataFrame(emp_1)

emp_2 = {'id': [10, 20, 33, 43, 53, 73],
       'name': ["A", "B", "H", "I", "J", 'L']}
emp_2_df = pd.DataFrame(emp_2)

pd.concat([emp_1_df, emp_2_df]).drop_duplicates() # UNION ALL
# To deduplicate things (equivalent of UNION), you’d also have to add .drop_duplicates().
```

# **CREATE TABLE / INSERT / UPDATE / DELETE**

```python
# CREATE TABLE Emp
# insert into Emp values (10, 'A')
# insert into Emp values (20, 'B')
# insert into Emp values (30, 'C')

emp = {'id': [10, 20, 30],
       'name': ["A", "B", "C"]}
emp_df = pd.DataFrame(emp) # Table is made

# After some time let's say we want to add do another 'insert into Emp values (40, 'D')'

emp_2 = {'id': [40], 'name':['D']}
emp_2_df = pd.DataFrame(emp_2)

emp = pd.concat([emp_df, emp_2_df])

emp_df
```

```python
# UPDATE Emp SET name = 'AAAA' WHERE id = 10 AND name = 'A'

emp_df.loc[(emp_df.id == 10) & (emp_df.name == 'A'), 'name'] = 'AAAA' # Happens in place

emp_df
```

```python
# DELETE FROM Emp WHERE id = 10 AND name = 'A'

emp_df.drop(emp_df[(emp_df.id == 10) & (emp_df.name == 'AAAA')].index)
```
