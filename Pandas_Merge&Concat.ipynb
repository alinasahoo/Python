**Merge**

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,4],"D":[50,60,70,80]})

var1
```

```python
var2
```

```python
pd.merge(var1,var2,on=["A"]) #joins the two tables on A being common
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,5],"D":[50,60,70,80]})

pd.merge(var1,var2,on=["A"]) #joins the two tables on A being common [inner join]
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,5],"D":[50,60,70,80]})

pd.merge(var1,var2,on=["A"],how="left") #joins the two tables on A being common [left join]
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,5],"D":[50,60,70,80]})

pd.merge(var1,var2,on=["A"],how="right") #joins the two tables on A being common [right join]
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,5],"D":[50,60,70,80]})

pd.merge(var1,var2,on=["A"],how="outer",indicator=True) #joins the two tables on A being common [outer join]
#indicator shows the merge col
```

```python
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=["_left","_right"])
```

**Concat**

```python
sr1 = pd.Series([1,2,3,4],index=["a","b","c","d"])
sr2 = pd.Series([5,6,7,8],index=["a","b","c","d"])

sr1
```

```python
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
pd.concat([var1,var2],axis=1)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,4],"D":[50,60,70,80]})

pd.concat([var1,var2],axis=0) #concats across rows
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2],"D":[50,60]})

pd.concat([var1,var2],axis=1) #concats across columns
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2],"D":[50,60]})

pd.concat([var1,var2],axis=1,join="inner")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,4],"D":[50,60,70,80]})

pd.concat([var1,var2],axis=0,keys=["var1","var2"])
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2 = pd.DataFrame({"A":[1,2,3,4],"D":[50,60,70,80]})

pd.concat([var1,var2],axis=1,keys=["var1","var2"])
```
