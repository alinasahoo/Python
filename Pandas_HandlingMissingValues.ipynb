Dropna and Fillna


```python
import pandas as pd

var = pd.read_csv("/content/data.csv")

var
```

```python
var.head()
```

```python
var.dropna() #drops the null values across the rows 18,22,28
```

```python
var.isnull() #checks if the values are null [indexed : 18, 22, 28] evaluates to true
```

```python
var.isnull().sum() #tells us which cols have what number of null values
```

```python
var.dropna(axis=0) #drops missing values along the rows [indexed 18,22,28]
```

```python
var.dropna(axis=1) #drop missing values across the columns [col named : date & calories]
```

```python
var.dropna(how="any") #drop missing values if any value is missing, rows 18,22,28
```

```python
var.dropna(how="all") #drop missing values if all values are missing, hence doesn't drop any

```

```python
var.dropna(subset=["Calories"]) #drops all the missing values only in the column named : Calories
```

```python
var.dropna(inplace=True) #drops the missing values in the dataframe

var
```

```python
import numpy as np
df = pd.DataFrame([[np.nan, 2, np.nan, 0], [3, 4, np.nan, 1],
                    [np.nan, np.nan, np.nan, 5]],
                  columns=list('ABCD'))
df
```

```python
df.dropna(thresh=2) #drops the rows which have less than 2 non-null values, index 2 has 1 non-null value [1]
```

```python
df.dropna(thresh=3)
#drops the rows which have less than 3 non-null values, index 0 has 2 non-null vales [2.0 and 0] & index 2 has 1 non-null value [1]
```
